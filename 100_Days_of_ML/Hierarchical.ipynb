```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Mall_Customers.csv')
```

```python
x=df.iloc[:,[2,3]].values
```

```python
import scipy.cluster.hierarchy as shc
d=shc.dendrogram(shc.linkage(x,method="ward"))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering as ac
hc=ac(n_clusters=5,linkage="ward")
y_pred=hc.fit_predict(x)
print(y_pred)
```

```python
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=100,color='blue',label='Cluster 1')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=100,color='red',label='Cluster 2')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=100,color='green',label='Cluster 3')
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],s=100,color='magenta',label='Cluster 4')
plt.scatter(x[y_pred==4,0],x[y_pred==4,1],s=100,color='black',label='Cluster 5')
plt.title('Customers')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Mall_Customers.csv')
```

```python
x=df.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans as km
wcss_list=[]
for i in range(1,11):
    k=km(n_clusters=i,init='k-means++',random_state=42)
    k.fit(x)
    wcss_list.append(k.inertia_)
plt.plot(range(1,11),wcss_list)
plt.show()
```

```python
from sklearn.cluster import KMeans as km
k=km(n_clusters=5,init='k-means++',random_state=42)
y_pred=k.fit_predict(x)
print(y_pred)
```

```python
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=100,color='blue',label='Cluster 1')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=100,color='red',label='Cluster 2')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=100,color='green',label='Cluster 3')
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],s=100,color='magenta',label='Cluster 4')
plt.scatter(x[y_pred==4,0],x[y_pred==4,1],s=100,color='black',label='Cluster 5')
plt.scatter(k.cluster_centers_[:,0],k.cluster_centers_[:,1],s=300,color='yellow',label='Centroid')
plt.title('Customers')
plt.legend()
plt.show()
```
